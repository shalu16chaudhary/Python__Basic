```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.isnull().sum()
```

```python
# Rows which has null values
z = titanic[titanic['age'].isnull()]
temp = z # to record a copy original nan age
print(z)
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
z = temp.index
titanic.iloc[z]
```
